## A. Build a baseline model (5 marks)

Let's download the data and read it into a <em>pandas</em> dataframe.

```python
concrete_data = pd.read_csv('concrete_data.csv')
concrete_data.head()
```

#### Let's check how many data points we have.

```python
concrete_data.shape
```

Let's check the dataset for any missing values.

```python
concrete_data.describe()
```

```python
concrete_data.isnull().sum()
```

The data looks very clean and is ready to be used to build our model.

#### Split data into predictors and target

The target variable in this problem is the concrete sample strength. Therefore, our predictors will be all the other columns.

```python
concrete_data_columns = concrete_data.columns

predictors = concrete_data[concrete_data_columns[concrete_data_columns != 'Strength']] # all columns except Strength
target = concrete_data['Strength'] # Strength column
```

Let's do a quick sanity check of the predictors and the target dataframes.

```python
predictors.head()
```

```python
target.head()
```

Let's save the number of predictors to *n_cols* since we will need this number when building our network.

```python
n_cols = predictors.shape[1] # number of predictors
```

## Create the Training and Test Datasets

```python
# import train_test_split, mean_squared_error from scikitlearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
X = predictors.values
y = target.values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=40)
print(X_train.shape); print(X_test.shape)
```

## Import Keras

#### Let's go ahead and install and import the Keras library

```python
# pip show tensorflow
```

```python
# pip show keras
```

```python
import keras
```

Let's import the rest of the packages from the Keras library that we will need to build our regression model.

```python
from keras.models import Sequential
from keras.layers import Dense
```

## Build a Neural Network

Let's define a function that defines our regression model for us so that we can conveniently call it to create our model.

```python
# define regression model
def regression_model():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=["mean_squared_error"])
    return model
```

## Train and Test the Network

Let's call the function now to create our model.

```python
# build the model
model = regression_model()
```

Next, we will train and test the model at the same time using the *fit* method.

```python
# fit the model
model.fit(X_train, y_train, epochs=50, verbose=2)
```

```python
pred_train= model.predict(X_train)
print(mean_squared_error(y_train,pred_train))

pred= model.predict(X_test)
print(mean_squared_error(y_test,pred))
```

```python
mean_squared_error_list = []

for i in range(50):
    model = regression_model()
    model.fit(X_train, y_train, epochs=50)
    pred_train= model.predict(X_train)
    pred= model.predict(X_test)

    mean_squared_error_list.append(mean_squared_error(y_test,pred))

print("Mean Squared Error List")
print(mean_squared_error_list)
print("Mean of Mean Squared Error List")
print(np.mean(mean_squared_error_list))
print("Standard Deviation of Mean Squared Error List")
print(np.std(mean_squared_error_list, ddof=1))
```

```python
print("Mean of Mean Squared Error List")
print(np.mean(mean_squared_error_list))
print("Standard Deviation of Mean Squared Error List")
print(np.std(mean_squared_error_list, ddof=1))
```

## B. Normalize the data (5 marks)

Normalize the data by substracting the mean and dividing by the standard deviation.

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
n_cols_norm = predictors_norm.shape[1] # number of predictors
```

## Create the Training and Test Datasets

## Build a Neural Network

## Train and Test the Network

```python
# build the model
model_norm = regression_model_norm()
```

Next, we will train and test the model at the same time using the *fit* method.

```python
# fit the model
model_norm.fit(X_norm_train, y_norm_train, epochs=50, verbose=2)
```

```python
pred_norm_train= model_norm.predict(X_norm_train)
print(mean_squared_error(y_norm_train,pred_train))

pred_norm= model_norm.predict(X_norm_test)
print(mean_squared_error(y_norm_test,pred_norm))
```

```python
mean_squared_error_list_norm = []

for i in range(50):
    model_norm = regression_model_norm()
    model_norm.fit(X_norm_train, y_norm_train, epochs=50, verbose=2)
    pred_norm_train= model_norm.predict(X_norm_train)
    pred_norm= model_norm.predict(X_norm_test)

    mean_squared_error_list_norm.append(mean_squared_error(y_norm_test,pred_norm))

print("Mean Squared Error List")
print(mean_squared_error_list_norm)
print("Mean of Mean Squared Error List")
print(np.mean(mean_squared_error_list_norm))
print("Standard Deviation of Mean Squared Error List")
print(np.std(mean_squared_error_list_norm, ddof=1))
```

## D. Increase the number of hidden layers (5 marks)

## Build a Neural Network

Let's define a function that defines our regression model for us so that we can conveniently call it to create our model.

```python
# define regression model
def regression_model_3layers():
    # create model
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols_norm,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error', metrics=["mean_squared_error"])
    return model
```

## Train and Test the Network

Let's call the function now to create our model.

```python
# build the model
model_3layers = regression_model_3layers()
```

Next, we will train and test the model at the same time using the *fit* method.

```python
# fit the model
model_3layers.fit(X_norm_train, y_norm_train, epochs=50, verbose=2)
```

```python
pred_3layers_train= model_3layers.predict(X_norm_train)
print(mean_squared_error(y_norm_train,pred_3layers_train))

pred_3layers= model_3layers.predict(X_norm_test)
print(mean_squared_error(y_norm_test,pred_3layers))
```

```python
mean_squared_error_list_3layers = []

for i in range(50):
    model_3layers = regression_model_3layers()    
    model_3layers.fit(X_norm_train, y_norm_train, epochs=50, verbose=2)
    pred_3layers_train= model_3layers.predict(X_norm_train)
    pred_3layers= model_3layers.predict(X_norm_test)

    mean_squared_error_list_3layers.append(mean_squared_error(y_norm_test,pred_3layers))

print("Mean Squared Error List")
print(mean_squared_error_list_3layers)
print("Mean of Mean Squared Error List")
print(np.mean(mean_squared_error_list_3layers))
print("Standard Deviation of Mean Squared Error List")
print(np.std(mean_squared_error_list_3layers, ddof=1))
```

```python
print("Mean of Mean Squared Error List")
print(np.mean(mean_squared_error_list_3layers))
print("Standard Deviation of Mean Squared Error List")
print(np.std(mean_squared_error_list_3layers, ddof=1))
```

The mean of the mean squared errors is lower than that of Step B.
